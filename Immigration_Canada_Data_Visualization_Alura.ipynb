# # Analyze immigration trends from Brazil to Canada from 1980 to 2013

```python
df.set_index('Country', inplace=True)
```

```python
df
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
brazil = df.loc['Brazil', years]
```

```python
brazil
```

```python
brazil_dic = {'year': brazil.index.tolist(), 'immigrants': brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dic)
```

```python
data_brazil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_brazil['year'], data_brazil['immigrants'])
plt.title('Migration from Brazil to Canada (1980-2013)')
plt.xlabel('Years')
plt.ylabel('Immigrants')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#creating figure

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(data_brazil['year'], data_brazil['immigrants'], lw=3, color='g')
ax.set_title('Immigration from Brazil to Canada\n (1980-2013)', fontsize = 18, loc='left')
ax.set_xlabel('Year', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.set_ylabel('Immigrants', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('brazil_immigration_canada.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].plot(data_brazil['year'], data_brazil['immigrants'])
axs[0].set_title('Migration from Brazil to Canada (1980-2013)')
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Immigrants')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)

axs[1].boxplot([data_brazil['immigrants']])
axs[1].set_title('Boxplot of Immigration from Brazil to Canada (1980-2013)')
axs[1].set_ylabel('Immigrants')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

# Plot of immigration from the 4 largest countries in South America to Canada (1980 to 2013)

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle('Immigration from the 4 largest countries in South America to Canada (1980 to 2013)')

axs[0,0].plot(df.loc['Brazil', years])
axs[0,0].set_title('Brazil')
axs[0,0].set_xlabel('Year')
axs[0,0].set_ylabel('Immigrants')
axs[0,0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0,0].grid(True)

axs[0,1].plot(df.loc['Colombia', years])
axs[0,1].set_title('Colombia')
axs[0,1].set_xlabel('Year')
axs[0,1].set_ylabel('Immigrants')
axs[0,1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0,1].grid(True)

axs[1,0].plot(df.loc['Argentina', years])
axs[1,0].set_title('Argentina')
axs[1,0].set_xlabel('Year')
axs[1,0].set_ylabel('Immigrants')
axs[1,0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1,0].grid(True)

axs[1,1].plot(df.loc['Peru', years])
axs[1,1].set_title('Peru')
axs[1,1].set_xlabel('Year')
axs[1,1].set_ylabel('Immigrants')
axs[1,1].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[1,1].grid(True)

ymin = 0
ymax = 7000
for ax in axs.flat:
    ax.set_ylim(ymin, ymax)

plt.tight_layout()
fig.savefig('immigration_4_south_america_canada.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.loc['Brazil', years], label='Brazil', color='g', lw=3)
plt.plot(df.loc['Colombia', years], label='Colombia', color='y', lw=3)
plt.plot(df.loc['Argentina', years], label='Argentina', color='b', lw=3)
plt.plot(df.loc['Peru', years], label='Peru', color='r', lw=3)

plt.xlabel('Year', fontsize=12)
plt.ylabel('Immigrants', fontsize=12)
plt.title('Immigration from 4 biggest country in South America to Canada(1980-2013)', fontsize=16)
plt.xticks(rotation=45)

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Analyzing all of south america

```python
df.head()
```

```python
america_south = df.query('Region == "South America"')
america_south
```

```python
america_south = america_south.rename(index={'Bolivia (Plurinational State of)': 'Bolivia'})
america_south = america_south.rename(index={'Venezuela (Bolivarian Republic of)': 'Venezuela'})
```

```python
america_south = america_south.sort_values(by='Total', ascending=True)
america_south
```

```python
colors = ['green' if country == 'Brazil' else 'silver' for country in america_south.index]

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(america_south.index, america_south['Total'], color=colors)
ax.set_title('Analysis of Immigration from Brazil to Canada in relation\nto other countries in South America (1980 to 2013)', fontsize=18, loc='left')
ax.set_xlabel('Total Immigrants', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_south['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=12, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', which='both', length=0)

plt.tight_layout()

fig.savefig('immigration_south_america_canada.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

# Top 10 countries with the most immigration to Canada


```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
top_10 = top_10.rename(index={'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom and Ireland'})
top_10
```

```python
sns.set_theme(style='darkgrid')
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=top_10, x='Total', y=top_10.index, orient= 'h', palette='tab10', hue=top_10.index, legend=False)
plt.xlabel('Total Immigrants', fontsize=14)
plt.ylabel('')
plt.title('Top 10 countries with the most immigration to Canada\n(1980-2013)', fontsize=18, loc='left')


plt.tight_layout()
plt.savefig('top_10_countries_immigration_canada.png', dpi=300, bbox_inches='tight', transparent=False)
plt.show()
```

# Creating an interactive graph with the countries of South America (plotly)

```python
america_south.head()
```

```python
df_america_south_clean = america_south.drop(columns=['Region', 'Continent', 'Total', 'DevName'])
df_america_south_final = df_america_south_clean.T
df_america_south_final
```

```python
df_america_south_final.columns = df_america_south_final.columns.astype(str)
```

```python
fig = px.line(df_america_south_final, x=df_america_south_final.index, y=df_america_south_final.columns, color='Country',
              title='Immigration from South America to Canada (1980-2013)', markers=True)

fig.update_layout(xaxis={'tickangle': -45},
                  xaxis_title='Year',
                  yaxis_title='Immigrants')

fig.show()
```
